### Diabetes Prediction

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../DataSet/diabetes.csv")
```

```python
data.sample(10)
```

```python
data.info()
```

```python
data.describe().T
```

```python
fig = plt.figure(figsize=(5,5))
for i in data.columns.drop("Outcome"):
    for j in data.columns.drop("Outcome"):
        if i != j:
            sns.scatterplot(x = i, y = j, data = data,hue = "Outcome", palette="dark",legend=True)
            plt.show()
```
